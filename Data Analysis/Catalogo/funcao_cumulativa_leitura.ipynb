```python
todas_as_datas = pd.date_range(start='2025-05-25', end='2025-06-05')
df_completo = pd.DataFrame({'data': todas_as_datas})
relato = df_completo.merge(relato, on='data', how='left')

relato

# Preencher campos ausentes
relato["tema"] = relato["tema"].fillna("Relato de um certo Oriente")
relato["tipo"] = relato["tipo"].fillna("Literatura")
relato["tipo_2"] = relato["tipo_2"].fillna("HASS")
relato["paginas"] = relato["paginas"].fillna(0).astype(int)

relato = relato.sort_values("data")
relato
```

```python
pontos = []
init = 0
for i, val in enumerate(relato['paginas'].values):
  init+=val
  tup = (int(init), i+1)
  pontos.append(tup)

x_vals = [y for x, y in pontos]
y_vals = [x for x, y in pontos]

# Criar o gráfico
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, marker='o', linestyle='-', color='blue', label='Cumulativo')
plt.title('Gráfico da Função Cumulativa')
plt.xlabel('DIas Leitura')
plt.ylabel('Valor Acumulado')
plt.grid(True)
plt.tight_layout()
plt.show()
```
