Separa√ß√£o de vari√°veis alvo e explicativas

```python
dados
```

```python
X = dados.drop('aderencia_investimento', axis=1)
y = dados['aderencia_investimento']
```

```python
X
```

```python
y
```

Transformando vari√°eis explicativas com One-Hot Enconding

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
colunas = X.columns
```

```python
one_hot = make_column_transformer((
    OneHotEncoder(drop = 'if_binary'),
    ['estado_civil', 'escolaridade', 'inadimplencia', 'fez_emprestimo']
    ),
    remainder = 'passthrough',
    sparse_threshold = 0)
```

```python
X_ = one_hot.fit_transform(X)
```

```python
one_hot.get_feature_names_out(colunas)
```

```python
pd.DataFrame(X_, columns = one_hot.get_feature_names_out())
```

Transoforma√ß√£o da Vari√°vel Alvo

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
y = label_encoder.fit_transform(y)
```

```python
y
```

Divis√£o de treino e teste

```python
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X_, y, stratify = y, random_state = 5)
```

Modelo de base
- dummy classifier (classifica√ß√£o a partir da categoria da vari√°vel alvo que tem maior frequ√™ncia, prevendo todos os valores com o mesmo resultado)

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy = DummyClassifier()
dummy.fit(X_treino, y_treino)

dummy.score(X_teste, y_teste)
```

√Årvore de Decis√£o

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore = DecisionTreeClassifier(random_state=5)
```

```python
arvore.fit(X_treino, y_treino)
```

```python
arvore.predict(X_teste)
```

```python
arvore.score(X_teste, y_teste)
```

Regra constru√≠da pela √°rvore

```python
from sklearn.tree import plot_tree
```

```python
nome_colunas = ['casado (a)',
                'divorciado (a)',
                'solteiro (a)',
                'fundamental',
                'medio',
                'superior',
                'inadimplencia',
                'fez_emprestimo',
                'idade',
                'saldo',
                'tempo_ult_contato',
                'numero_contatos']
```

```python
plt.figure(figsize=(15,6))
plot_tree(arvore, filled=True, class_names = ['n√£o', 'sim'], fontsize=1, feature_names=nome_colunas);
```

```python
arvore.score(X_treino, y_treino)
```

```python
arvore = DecisionTreeClassifier(random_state=5, max_depth=3)
arvore.fit(X_treino, y_treino)
```

```python
arvore.score(X_treino, y_treino)
```

```python
arvore.score(X_teste, y_teste)
```

```python
plt.figure(figsize=(15,6))
plot_tree(arvore, filled=True, class_names = ['n√£o', 'sim'], fontsize=10, feature_names=nome_colunas);
```

Normaliza√ß√£o

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
norm = MinMaxScaler()
X_treino = norm.fit_transform(X_treino)
```

```python
pd.DataFrame(X_treino)
```

Usando KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_treino, y_treino)
```

```python
X_teste_norm = norm.transform(X_teste)
```

```python
knn.score(X_teste_norm, y_teste)
```

Melhor modelo

```python
print(f'Acur√°cia Dummy: {dummy.score(X_teste, y_teste)}')
print(f'Acur√°cia Dummy: {arvore.score(X_teste, y_teste)}')
print(f'Acur√°cia Dummy: {knn.score(X_teste_norm, y_teste)}')
```
