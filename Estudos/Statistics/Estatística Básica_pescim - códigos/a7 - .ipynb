```python
import math
import numpy as np
import statistics as stats
```

```python
def vari(a):
  media = sum(a)/len(a)
  sum_ = 0
  for i in a:
    sum_ += (i-media)**2

  variance = sum_/(len(a)-1)
  desv = math.sqrt(variance)

  return variance
```
