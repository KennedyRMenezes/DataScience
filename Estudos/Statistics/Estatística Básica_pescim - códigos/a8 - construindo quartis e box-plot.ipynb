```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
observacoes = [76, 120, 135, 149, 157, 163, 171, 178, 190, 207, 87, 121, 141, 150, 158, 163, 171, 180,
               193, 208, 97, 123, 142, 150, 158, 165, 172, 180, 194, 218, 101, 131, 143, 151, 158, 167,
               174, 181, 196, 221, 105, 133, 145, 153, 158, 167, 174, 181, 199, 228, 110, 133, 146, 154,
               160, 168, 175, 183, 199, 229, 115, 134, 148, 154, 160, 169, 176, 184, 200, 237, 118, 135,
               149, 156, 160, 170, 176, 186, 201, 245]

data = np.sort(observacoes)

fig, ax = plt.subplots(figsize=(5, 7))
box = ax.boxplot(data, vert=True, patch_artist=True)

q1 = box['whiskers'][0].get_ydata()[1]  # Primeiro quartil (Q1)
q3 = box['whiskers'][1].get_ydata()[1]  # Terceiro quartil (Q3)
min_val = box['caps'][0].get_ydata()[0]  # Valor mínimo sem outlier
max_val = box['caps'][1].get_ydata()[0]  # Valor máximo sem outlier
median = box['medians'][0].get_ydata()[0]  # Mediana (Q2)
outliers = box['fliers'][0].get_ydata()  # Outliers

x_pos = 1.1  # Posição X para os textos
# ax.text(x_pos, min_val, f"Min: {min_val:.2f}", fontsize=10, verticalalignment='center')
ax.text(x_pos, q1, f"Q1: {q1:.2f}", fontsize=10, verticalalignment='center')
ax.text(x_pos, median, f"Mediana: {median:.2f}", fontsize=10, verticalalignment='center')
ax.text(x_pos, q3, f"Q3: {q3:.2f}", fontsize=10, verticalalignment='center')
# ax.text(x_pos, max_val, f"Max: {max_val:.2f}", fontsize=10, verticalalignment='center')

# Outliers (se houver)
for outlier in outliers:
    ax.text(x_pos, outlier, f"{outlier:.2f}", fontsize=10, verticalalignment='center', color='red')

ax.set_xticks([])
ax.set_title("Boxplot com valores")
plt.show()
```
