```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Definição dos parâmetros
pastilhas = np.arange(0, 4, 1)  # Sequência de 0 a 3
prob_suc = 0.8
size = 3  # Número de tentativas

# Cálculo da distribuição binomial
probabilidades = binom.pmf(pastilhas, size, prob_suc)

# Plotage
plt.figure(figsize=(7,5))
plt.stem(pastilhas, probabilidades, linefmt='b-', markerfmt='bo', basefmt="gray")
plt.xlabel("Número de Pastilhas que Passam no teste")
plt.ylabel("Probabilidades")
plt.xticks(pastilhas)
plt.yticks(np.arange(0, 0.5, 0.05))
plt.axhline(y=0, color='gray', linewidth=2.5)  # Linha horizontal em y=0
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
