```python
# Gerando dados aleatrios para os trs conjuntos
np.random.seed(42)
dados1 = np.random.normal(50, 10, 100)  # Mdia 50, desvio padro 10
dados2 = np.random.normal(60, 15, 100)  # Mdia 60, desvio padro 15
dados3 = np.random.normal(70, 20, 100)  # Mdia 70, desvio padro 20

# Criando o boxplot
plt.boxplot([dados1, dados2, dados3], patch_artist=True,
            boxprops=dict(facecolor="lightblue", color="black"),
            medianprops=dict(color="red", linewidth=2))

# Adicionando rtulos
plt.xticks([1, 2, 3], ["Grupo 1", "Grupo 2", "Grupo 3"])
plt.title("Comparao de Trs Boxplots")
plt.ylabel("Valores")

# Exibir o grfico
plt.show()
```
