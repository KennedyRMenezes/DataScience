42 - MatplotLib Part1

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
x
```

```python
y
```

FUNCTIONAL METHOD TO PLOT

```python
plt.plot(x, y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1,2,1) # plt.subplot(rows, cols, position or which plot is reffering to)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

OBJECT ORIENTED API METHOD

Instantiate figure objects and then call methods or attributes from that object.

```python
fig = plt.figure() #imaginary blank canvas

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()

#manually choose axes
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("larger plot")


axes2.plot(y,x)
axes2.set_title("smaller plot")
```

43 - MatplotLib Part 2

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

Fig Size and DPI

```python
fig = plt.figure(figsize=(8,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2), dpi=100)

axes[0].plot(x,y)

axes[1].plot(x,y)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="X Squared")
ax.plot(x,x**3, label="X Cubed")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")

ax.legend()
```

44 - MatplotLib - Part 3 - Plot appearence

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=3,alpha=0.5, ls='--', marker=".", markersize=20,
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor="green")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=3, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
