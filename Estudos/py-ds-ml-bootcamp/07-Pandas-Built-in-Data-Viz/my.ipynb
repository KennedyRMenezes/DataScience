```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
df1 = pd.read_csv("df1", index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv("df2")
```

```python
df2.head()
```

```python
df1['A'].hist(bins=20)
```

```python
df1["A"].plot(kind='hist')
```

```python
df1["A"].plot.hist()
```

```python
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

```python
df1.plot.scatter(x='A',y="B",c='C',cmap="coolwarm")
```

```python
df1.plot.scatter(x='A',y="B",s=df1['C']*100,edgecolor='black',lw=0.7)
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm',edgecolor='white',lw=0.1)
```
