```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

With Principle Component Analysis is find out waht components are the most importante ones that explains the most variance of this data set.

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

- Instantiate a PCA Object

- Find the Principal Components using the fit method and then 

- Apply the rotation in dimensionality reduction by calling transform.

```python
# PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1],c=cancer['target'],cmap='coolwarm',alpha=0.7)
plt.xlabel("First Princple Component")
plt.ylabel("Second Principle Component")
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='coolwarm')
```
